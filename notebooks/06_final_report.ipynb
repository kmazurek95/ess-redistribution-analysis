5. **AI exposure:** Direct effect is null (p = 0.857); indirect channel through inequality identified

### REPAIR relevance

- Documents institutional mediation of economic effects on preferences

## 3. Theoretical Contributions

1. **Social exposure framework:** Technological threat is not purely technical - political impact depends on institutional context. Social Exposure = Technical Risk x Institutional Weakness.
2. **Institutional buffering:** Strong institutions dampen technological threat effects through collective protection. Policy implication: institution-building over compensation alone.
3. **Regime-specific AI politics:** No universal response to automation. Path dependency shapes how each regime type processes technological disruption.
4. **Indirect channel:** AI affects redistribution preferences not through direct perceived threat but through reshaping income distributions and inequality.

## 5. Methodological Notes

- **Multilevel models:** REML estimation via statsmodels MixedLM. Random intercepts for country; random slopes tested for income and ideology.
- **Centering:** Grand-mean centering for Level-1, z-score standardization for Level-2.
- **Typologies:** Esping-Andersen 5-regime and VoC 4-type classifications. Social exposure composite constructed as AI x institutional weakness.
- **Data integration:** ESS Round 9 merged with World Bank/OECD (economic), ICTWSS (unions), and Felten AIOE (AI exposure).
- **Reproducibility:** Modular `src/` code, `config.py` for paths and parameters, automated `create_analysis_dataset()` pipeline.

## 6. Future Directions

## 7. Policy Implications

- **Liberal regimes (GB, IE):** High AI exposure + weak institutions = high vulnerability. Priority: strengthen labor market institutions, expand active labor market policies.
- **Social Democratic (Nordics):** Moderate exposure, strong buffers. Priority: maintain institution strength, invest in lifelong learning, adapt collective bargaining to new sectors.
- **Continental (DE, FR, etc.):** Moderate exposure, labor market dualism risk. Priority: reduce insider/outsider gaps, portable benefits, extend co-determination to AI deployment.
- **Post-Communist (Eastern Europe):** Variable exposure, weak institutions. Priority: build union capacity, targeted regional support, EU coordination.
- **Cross-cutting:** Worker participation in AI governance, progressive taxation of AI productivity gains, focus on AI-complementary skills.

## 8. Limitations

- **Cross-sectional:** Cannot establish causality; need panel data or experiments
- **AI measurement:** Country-level aggregates, not individual exposure; Felten AIOE captures cognitive tasks only
- **Coverage:** European countries only; no US, Canada, Australia for Liberal regime comparison
- **Missing institutional data:** Non-OECD countries lack some indicators
- **Omitted variables:** Cultural factors, party systems, media environment not modeled
- **Preference endogeneity:** Preferences may be shaped by institutions rather than exogenous
