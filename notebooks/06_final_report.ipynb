## 1. Executive Summary

### Research Question

**How do individual characteristics, national economic context, welfare regime institutions, and AI/automation exposure shape public support for government redistribution across European welfare states?**

### Data & Methods

- **Sample:** European Social Survey Round 9 (2018), ~36,000 respondents, 28 countries
- **Outcome:** Support for government redistribution (1-5 scale)
- **Methods:** Multilevel regression models (random intercepts and slopes)
- **Levels:** Individual characteristics (Level-1), country context (Level-2)
- **Extensions:** Welfare regime typologies, institutional mediation, AI exposure

### Key Findings

**1. Substantial cross-national variation (ICC ≈ 10-15%)**
- Redistribution preferences vary significantly across countries
- Justifies multilevel approach and comparative focus

**2. Individual-level effects (consistent across countries)**
- Income: Negative effect (self-interest)
- Ideology: Strong negative effect (political values)
- Trust: Positive effect (state capacity beliefs)
- Meritocracy: Negative effect (deservingness beliefs)

**3. Country-level economic context matters**
- Gini coefficient: Mixed effects (demand vs. power resources)
- GDP per capita: Negative effect (affluence reduces demand)
- Unemployment: Positive effect (insecurity increases demand)

**4. Welfare regimes structure preferences**
- Social Democratic regimes: Highest support
- Liberal regimes: Lowest support
- Institutions moderate individual-level effects

**5. AI/automation exposure reshapes politics**
- Higher AI exposure → more redistribution support
- Effect strongest where institutions are weak (Liberal regimes)
- "Social exposure" (AI × weak institutions) predicts preferences

This analysis directly addresses REPAIR's core questions:

## 3. Theoretical Contributions

### Extending Classic Welfare State Theory

**Esping-Andersen (1990) → AI Era:**
- Classic typology still captures real variation
- But technological change adds new dimension
- Regimes respond differently to AI disruption

**Power Resources Theory (Korpi 1983) → Digital Age:**
- Union density still matters for preferences
- But AI threatens traditional labor organization
- New forms of collective action may be needed

**Varieties of Capitalism (Hall & Soskice 2001) → Automation:**
- CME/LME distinction relevant for AI politics
- Coordinated economies better positioned to manage disruption
- Institutional complementarities extend to technological change

### New Theoretical Contributions

**1. Social Exposure Framework**
- Technological threat is not purely technical
- Political impact depends on institutional context
- Formula: Social Exposure = Technical Risk × Institutional Weakness

**2. Institutional Buffering Hypothesis**
- Strong institutions dampen technological threat effects
- Mechanism: Collective protection reduces individual vulnerability
- Policy implication: Institution-building > compensation alone

**3. Regime-Specific AI Politics**
- No universal response to automation
- Liberal regimes: Market + minimal safety net
- Social Democratic: Active labor market + strong protection
- Continental: Insurance + moderate protection
- Path dependency shapes technological politics

## 5. Methodological Contributions

### Multilevel Modeling Approach

**Why multilevel models?**
- Individuals nested within countries
- Standard errors clustered by country
- Allows decomposition of variance (ICC)
- Enables cross-level interactions

**Technical implementation:**
- REML estimation (pymer4 + R lme4)
- Random intercepts for country variation
- Random slopes for heterogeneous effects
- Robust standard errors

### Variable Transformations

**Level-1 (Individual): Grand-mean centering**
- Interpretation: Effect at the mean of predictor
- Allows meaningful intercept interpretation
- Separates within- and between-country effects

**Level-2 (Country): Z-score standardization**
- Effect sizes comparable across predictors
- Mean = 0, SD = 1
- One-unit change = one standard deviation

### Typology Construction

**Esping-Andersen (5 regimes):**
- Theoretically grounded classification
- Captures welfare state variation
- Enables regime-specific analysis

## 7. Policy Recommendations

### For Liberal Regimes (UK, Ireland)

**Problem:** High AI exposure + weak institutions = high vulnerability

**Recommendations:**
1. **Strengthen labor market institutions**
   - Promote collective bargaining
   - Extend worker representation to gig economy
   - Increase union density through policy support

2. **Expand active labor market policies**
   - Training and reskilling programs
   - Job search assistance for displaced workers
   - Wage subsidies for transitioning occupations

3. **Universal basic income / negative income tax**
   - Cushion against AI-driven job losses
   - Reduce means-testing stigma
   - Pilot programs in high-exposure regions

### For Social Democratic Regimes (Nordics)

**Problem:** Moderate AI exposure, but institutions may not be sufficient

**Recommendations:**
1. **Maintain strong institutions**
   - Preserve union density (prevent decline)
   - Adapt collective bargaining to new sectors
   - Ensure AI-exposed workers are covered

2. **Invest in education and lifelong learning**
   - Early childhood education (reduce skill gaps)
   - Adult education subsidies
   - Digital literacy programs

3. **Proactive industrial policy**
   - Steer AI development toward job creation
   - Public investment in AI-complementary sectors
   - Worker participation in AI governance

### For Continental Regimes (Germany, France, etc.)

**Problem:** Moderate AI exposure, but labor market dualism

**Recommendations:**
1. **Reduce labor market dualism**
   - Extend protections to temporary workers
   - Reduce EPL gap between regular and temporary
   - Prevent AI-driven expansion of precarious work

2. **Reform social insurance**
   - Portable benefits (not tied to single employer)
   - Universal coverage (including self-employed)
   - Adapt to non-standard work arrangements

3. **Works councils and co-determination**
   - Extend worker voice to AI implementation
   - Negotiate AI deployment terms
   - Share productivity gains

### For Post-Communist Regimes (Eastern Europe)

**Problem:** Variable AI exposure, weak institutions, transition effects

**Recommendations:**
1. **Strengthen nascent institutions**
   - Build union capacity (currently low)
   - Establish collective bargaining frameworks
   - Learn from Western European experiences

2. **Targeted regional policies**
   - High AI exposure regions need extra support
   - Industrial transition assistance
   - Infrastructure investment

3. **EU-level coordination**
   - Access to structural funds

## 8. Limitations and Extensions

### Data Limitations

**1. Cross-Sectional Design**
- Cannot establish causality definitively
- Correlation ≠ causation
- Need: Panel data or experiments

**2. AI Exposure Measurement**
- Country-level aggregates (not individual exposure)
- Assumes uniform exposure within countries
- Need: Occupation-level linkage

**3. ESS Limitations**
- Self-reported preferences (not behavior)
- Social desirability bias possible
- Survey mode effects (face-to-face vs. online)

**4. Country Coverage**
- Limited to European countries
- No US, Canada, Australia (for Liberal regime comparison)
- No Asian or Latin American countries

**5. Missing Data on Institutions**
- Non-OECD countries lack some indicators
- Limits institutional analysis to 21 countries
- Need: Alternative data sources for Balkans

### Methodological Limitations

**1. Multilevel Models Assume:**
- Linear relationships (may not always hold)
- Normally distributed errors
- No measurement error in predictors

**2. Typology Choices**
- Esping-Andersen is one of many possible classifications
- Some countries are ambiguous (e.g., Switzerland)
- Regime change over time not modeled

**3. Omitted Variables**
- Cultural factors (values, norms)
- Political factors (party systems, electoral rules)
- Media and information environment

**4. Aggregation Bias**
- Country-level variables may mask within-country variation
- Ecological fallacy concerns
- Need: More granular geographic data

### Theoretical Limitations

**1. Preference Formation**
- Assumes preferences are stable and exogenous
- In reality: Preferences may be endogenous to institutions
- Need: Dynamic models of preference change

**2. Political Economy Mechanisms**
- Preferences ≠ policies
- Need to model: Voting, party competition, coalitions
- Link: Mass preferences → political outcomes

**3. AI as a "Black Box"**
- Assumes AI is a uniform technological shock
- In reality: Many different AI technologies
- Need: Disaggregate by AI type (generative, analytical, robotic)

### Extensions to Address Limitations

**1. Panel Data Analysis**
- ESS Rounds 1-10 (2002-2020)
- Within-country variation over time
- Fixed effects models

**2. Individual-Level AI Exposure**
- Link respondent occupation to AI probabilities
- Test individual-level mechanisms
- Heterogeneous effects by worker type

**3. Survey Experiments**
- Randomize AI threat framing
- Causal identification
- Test specific mechanisms

**4. Qualitative Methods**
- Case studies of specific countries
- Elite interviews
- Process tracing

**5. Comparative Expansion**
- Include non-European countries
- Test generalizability
- Different welfare regime types
